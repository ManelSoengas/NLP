L'objectivo es traducir los tuits a un tercer idioma y de éste al original. De esta manera podemos aumentar las muestras de tuits de un idioma dado:



1.   Español------>Alemán------>Español
2.   Inglés------->Alemán------>Inglés

La nuevas muestras tienen diferencias en relación a los tuits originales. De esta manera podemos duplicar las muestras y entrenar el modelo con más ejemplos.
Aún así el conjunto de muestras tendran la clase objetivo desequilibrada.

```python
### Load dependences
from googletrans import Translator
from mtranslate import translate
import time
import pandas as pd

# translator = Translator()
```

```python
#### Data Path
PathDataSet = "../content/drive/MyDrive/Code/EXITS/Data/"
FileName = "EXIST2021_training.tsv"
#### Load tsv as a Data Frame
df_train = pd.read_csv(PathDataSet + FileName, sep='\t')

#### See the Data
df_train.head()
```

```python
df_train = df_train[df_train['language'] == 'en']  # Filtrar las filas con language='es'
df_train.reset_index(drop=True, inplace=True)  # Restablecer el índice y descartar el índice anterior

#### See the Data
df_train.head()
```

```python
df_train.shape
```

```python
### Create Columns for all comments in English after translation
df_train['German'] = ''
df_train.head()
```

```python
### Fill the columns 'English' with Spanish comments translated to English and original English comments
for index in range(0, 3541, 50):
  from googletrans import Translator
  translator = Translator()
  df_train.loc[index:index+51, ['German']] = df_train.loc[index:index+51].apply(lambda x : translator.translate(x['text'], src='en', dest='de').text if x['language']=='en' else x['text'] ,axis=1)
  time.sleep(1)
  # 6 min
```
