```python
# Importando las bibliotecas necesarias
import gensim
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from gensim.models import Word2Vec

# Preparando un conjunto de datos de ejemplo
ejemplo_oraciones = [
    "La inteligencia artificial es fascinante",
    "La educación y la tecnología están estrechamente relacionadas",
    "El alpinismo requiere resistencia y fuerza",
    "La lectura enriquece el conocimiento",
    "La escalada es un deporte emocionante",
    "La inteligencia artificial tiene un gran impacto en la educación"
    "El futbol es considerado un deporte de masas"
    "Futbol, escalada y alpinismo son deportes"
]

# Procesando el texto
ejemplo_procesado = [oracion.lower().split() for oracion in ejemplo_oraciones]

# Entrenando un modelo Word2Vec
modelo = Word2Vec(ejemplo_procesado, vector_size=100, window=5, min_count=1, workers=4)

# Palabras para visualizar
palabras = ['inteligencia', 'educación', 'tecnología', 'alpinismo', 'lectura', 'escalada', 'futbol', 'deporte']

# Extrayendo los vectores de estas palabras
vectores = [modelo.wv[palabra] for palabra in palabras if palabra in modelo.wv]

# Usando PCA para reducir la dimensionalidad
pca = PCA(n_components=2)
resultado_pca = pca.fit_transform(vectores)

# Visualización
plt.figure(figsize=(10, 6))
plt.scatter(resultado_pca[:, 0], resultado_pca[:, 1])
for i, palabra in enumerate(palabras):
    plt.annotate(palabra, xy=(resultado_pca[i, 0], resultado_pca[i, 1]))
plt.title("Visualización de Embeddings de Palabras con PCA")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.grid(True)
plt.show()
```
